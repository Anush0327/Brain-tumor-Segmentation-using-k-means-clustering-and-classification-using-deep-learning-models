```python
from tensorflow import keras
from sklearn import metrics
```

```python
import pathlib
data_dir=r'C:\Akash\Desktop\Dataset'
data_dir=pathlib.Path(data_dir)
data_dir

images_dict = {
    'glioma': list(data_dir.glob('glioma/*')),
    'meningioma': list(data_dir.glob('meningioma/*')),
    'pituitary': list(data_dir.glob('pituitary/*')),
    'notumor': list(data_dir.glob('notumor/*')),
}

labels_dict={
    'notumor':0,
    'glioma':1,
    'meningioma':2,
    'pituitary':3,
}

import cv2
X, Y = [], []

for image_type, images in images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(224,224))
        X.append(resized_img)
        Y.append(labels_dict[image_type])

#converting into numpy
import numpy as np
X=np.array(X)
Y=np.array(Y)

#normalisation
X=X/255
Y=Y

import cv2
import numpy as np
from sklearn.cluster import KMeans
X1=[]

for image in X:
    image_reshaped = image.reshape(image.shape[0]*image.shape[1], image.shape[2])
    X1.append(image_reshaped)
print("Shape of original image {}".format(X[0].shape))
print("Shape of reshaped image {}".format(X1[0].shape))
def kmeans_over_image(number_of_clusters, org_image, reshaped_image):
    kmeans = KMeans(n_clusters=number_of_clusters, random_state=0)
    kmeans = kmeans.fit(reshaped_image)
    image_cluster = kmeans.cluster_centers_[kmeans.labels_]
    # reshaping image size form 2D to 3D
    image_cluster = image_cluster.reshape(org_image.shape[0], org_image.shape[1], org_image.shape[2])

    return image_cluster
X2 = []

for i in range(len(X)):
    X2.append(kmeans_over_image(4, X[i], X1[i]))
    print(i)

X2=np.array(X2)
X2.shape
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X2, Y,test_size=0.3, random_state=0)
```

```python
vgg19=keras.models.load_model("Model.h5")
y_pred=vgg19.predict(X_test)
```

```python
temp=[]
for i in range(0,1054):
    temp.append(np.argmax(y_pred[i]))
print(temp)
cm=metrics.confusion_matrix(y_test,temp)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = labels_dict) 
import matplotlib.pyplot as plt 
cm_display.plot()
plt.show() 
from sklearn.metrics import classification_report
predictions=np.argmax(y_pred,axis=1)
predictions = predictions.reshape(1,-1)[0]

print(classification_report(y_test, predictions, target_names = ['no tumor (Class 0)','glioma (Class 1)','meningioma (Class 2)','pitutary (Class 3)']))
```
